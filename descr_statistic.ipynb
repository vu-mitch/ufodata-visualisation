### 2.a) Distribution of Data
***
What is the relative frequency of different day time periods?<br>
Instead of only seperating times into day and night we went into depth-analysis of times, just for the sake of a beautiful pie plot<br>

- midnight - 6am
- 6am - noon
- noon - 6pm
- 6pm - midnight

```python
# filter day times of occured ufo sightings and NAN values
list_times = []
for x in df_alien["date_occured"]:
    if isinstance(x, str):
        list_times.append(x[-8:])
```

```python
# filter data after day periods
import re

midnight_6am = []
six_am_noon = []
noon_6pm = []
six_pm_midnight = []

for x in sorted(list_times):
    if re.findall(r"^0[0-5]", x):
        midnight_6am.append(x)
    if re.findall(r"^0[7-9]", x) or re.findall(r"^1[0,1]", x):
        six_am_noon.append(x)
    if re.findall(r"^1[2-7]", x):
        noon_6pm.append(x)
    if re.findall(r"^1[8,9]", x) or re.findall(r"^2[0-3]", x):
        six_pm_midnight.append(x)
        
```

#### Usage of pie chart
- the pie chart is the best option for here, since the whole data can be described by 4 different categories, and we want to know the relative frequency of each distribution

```python
labels = "midnight-6am", "6am-noon", "noon-6pm", "6pm-midnight"
sizes = [len(midnight_6am), len(six_am_noon), len(noon_6pm), len(six_pm_midnight)]
explode = (0.05, 0.05, 0.05, 0.05)
colors = ['#a903fc','#66b3ff','#1403fc','#5040ff']

plt.figure(figsize=(16,9))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, pctdistance=0.8, colors=colors, textprops={'fontsize': 24})

#draw circle
centre_circle = plt.Circle((0,0),0.60,fc='black')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("During which day period are most ufo sightings?", fontsize=25)
plt.tight_layout()
plt.legend(loc="upper right", fontsize=13)
plt.show()
```

### 2.c) Mean
***
- What is the average time of ufo sightings?

#### Calculating the average time of ufo sightings
How do we calculate the average of day times?
- Best solution: clock times represented as radians, 24h represented on unit circle,  usage of trigonometry and complex numbers <br>
<br>

#### Represent times on unit circle
Start at midnight going counterclockwise <br>
- north / pi/2 : 6am<br>
- east / 0 or 2pi : midnight<br>
- south / 3pi/2 : noon<br>
- west / pi : 6pm<br>


#### First step: get an idea over which period of day time could lie the average = represent all times on unit circle

implemented functions:
- function which returns a list of radians
- function which calculates the mean

```python
# calulate the mean of angles and convert time into radians
def mean_angles(angles: list):
    radians =  [math.radians(a) for a in angles]
    radius = 1 # unit circle
    # turn into polar representation and calculate mean between complex numbers
    polar = sum(radius * (math.cos(r) + math.sin(r) * 1j) for r in radians) / len(angles)
    # get phase of complex number in order to obtain angle
    phase = math.atan2(polar.imag, polar.real)
    mean_angle = math.degrees(phase)
    return mean_angle, radians
```

```python
def mean_time(times: list):
    t = [time.split(":") for time in times]
    # convert times in seconds
    seconds = [int(h) * 3600 + int(m) * 60 + int(s) for h, m, s in t]
    daily_sec = 24 * 60 * 60
    # convert time to angles: ratio (sec/daily_sec) * 360 degrees
    to_angles = [sec / daily_sec * 360 for sec in seconds]
    mean_of_angles, radians = mean_angles(to_angles)
    # convert degrees in seconds
    # needed for calulating standard deviation
    mean_in_seconds = mean_of_angles * daily_sec / 360
    # handle negative times by adding a whole day
    if mean_in_seconds < 0:
        mean_in_seconds += daily_sec
    
    h, m = divmod(mean_in_seconds, 3600)
    m, s = divmod(m, 60)
    time = ("{:02d}:{:02d}:{:02d}".format(int(h), int(m), int(s)))
    return time, radians, mean_in_seconds, seconds
```

```python
# convert radians to polar form: recall z = r * (cos(phi) + i * sin(phi))
x = np.cos(radians)
y = np.sin(radians)
```

```python
#figure
plt.figure(figsize=(10,10))

#draw circle
circle = plt.Circle((0, 0), 1, color='white', fill=False, clip_on=False)

ax = plt.gca()
ax.add_patch(circle)
plt.axis('scaled')

#add times 
def draw_clock(a, b):
    plt.plot([0, a], [0, b], color='blue', linewidth=.05, alpha=0.4)

for i in range(len(radians)):
    draw_clock(x[i], y[i])
    
#add hours on face
ax.text(1.02, 0, r'midnight', fontsize=20)
ax.text(-0.07, 1.03, r'6am', fontsize=20)
ax.text(-1.20, 0, r'noon', fontsize=20)
ax.text(-0.07, -1.08, r'6pm', fontsize=20)

#final clock
plt.title("All times converted to radians on unit circle", fontsize=15)
plt.axis('off')
plt.show()
```

###  2.d) Standard Deviation
***
- How far are times of ufo entries spread from the average?
- Steps: implement function which takes list of day times converted into seconds and the mean in seconds

```python
# return in seconds
def std_dev(array: list, mean: int):
    n = len(array)
    deviations = [(x - mean) ** 2 for x in array]
    variance = sum(deviations) / n
    std_dev = np.sqrt(variance)
    return std_dev

std_dev = std_dev(seconds, mean_in_seconds)
```

```python
# convert seconds into hours
h, m = divmod(std_dev, 3600)
m, s = divmod(m, 60)
```
